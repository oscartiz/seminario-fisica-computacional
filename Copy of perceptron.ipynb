#**Importar librerias necesarias de python**

```python
import numpy as np
import matplotlib.pyplot as plt
```

#**Definir muestra de entrenamiento**

```python
x = np.array([[0, 0, 0], [1, 0, 1], [1, 1, 0], [0, 1, 1]])
```

```python
plt.scatter(x[:,0],x[:,1],s=100.0,c=x[:,2])
plt.title("Visualizacion de la logica AND")
plt.xlabel("X1")
plt.ylabel("X2")
```

#**Definimos la funcion perceptron**

```python
def perceptron(data,weights,bias):
  y_temp = (data[0]*weights[0])+(data[1]*weights[1])+bias 
 # print("y_temp ",y_temp)
  return 1.0 if y_temp >=0 else 0.0
```

```python
perceptron([1,0],[0.2,0.5],-1)
```

##**inicia proceso de entrenamiento**

```python
def training(training_data,l_rate):
    #inicializamos los pesos y el valor del bias de forma aleatoria entre 0 y 1 antes del ciclo de entrenamiento
    weights = [ np.random.uniform(0,1) for i in range(len(training_data[0])-1)]
    bias = np.random.uniform(0,1)
    print("initial weights  ",weights)
    sum_error = 9999
    epoch = 0 
    error_grafica = np.array([])
    epoch_grafica = np.array([])
    # el algoritmo va a seguir iterando hasta que el error sea cero (lo que indica que aprendio la logica AND) 
    while sum_error != 0.0:
      sum_error = 0.0
      for row in training_data:
        error = 0.0
        prediction = perceptron(row,weights,bias)
        # error que produce la comparacion de cada data de entrenamiento y el valor obtenido por el perceptron 
        error = row[2] - prediction
        sum_error+=pow(error,2)
        print("valor real", row[2]," prediction  ",prediction," error  ",error," sum error  ", sum_error)
        if error != 0.0:
          bias = bias + l_rate*error
          for i in range(len(row)-1):
            weights[i] = weights[i] + l_rate*error*row[i]
            print(" new weight[",i,"] = ",weights[i])

        # para cada iteracion en los datos graficamos la linea recta que resulta de los pesos y bias
        plot_x = np.array([np.min(x[:, 0] - 0.2), np.max(x[:, 1]+0.2)])
        plot_y = - 1 / weights[0] * (weights[1] * plot_x+bias)
        plot_y = np.reshape(plot_y, [2, -1])
        plot_y = np.squeeze(plot_y)

        plt.scatter(x[:,0],x[:,1],s=100.0,c=x[:,2])
        plt.plot(plot_x, plot_y, color='k', linewidth=2)
        plt.xlim([-0.2, 1.2]); plt.ylim([-0.2, 1.25]);
        plt.show()

      # cada iteracion en los datos produce un nuevo epoch
      error_grafica = np.append(error_grafica, sum_error)
      epoch = epoch + 1
      epoch_grafica = np.append(epoch_grafica, epoch)

      print("Iteration # ",epoch," Error = ",sum_error)
      print("==================================================")
      plt.scatter(epoch_grafica, error_grafica)
```

```python
training(x,0.1)
```
