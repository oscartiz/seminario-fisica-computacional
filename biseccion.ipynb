```python
from math import *
def f(x):
  return x**2-2;
def biseccion (a,b,e):
  m=b;
  if (f(a)*f(b)>0):
    print ('la funcion no cambia de signo')
  
  while (abs((a-b)/m)>e):
    m=(a+b)/2;
    if (f(a)*f(m)<0):
      b=m;
    if (f(m)*f(b)<0):
      a=m;
  print('x','=',m)

biseccion(1,2,10**-10)
```
