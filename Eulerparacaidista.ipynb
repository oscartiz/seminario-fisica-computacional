```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
def f(t,v):
    return g-c*v/m
```

```python
c=symbols('c') #constante
t=symbols('t') #Variable para guardar 
m=symbols('m') #Masa
g=symbols('g') #gravedad
V=Function('V') #Funcion velocidad en funcion de tiempo
```

```python
Eq(V(t).diff(),g-c*V(t)/m)
```

```python
ED=Eq(V(t).diff(),g-c*V(t)/m)
dsolve(ED,V(t),ics={V(0):0}) #Solucion analitica de la ecuacion diferencial
```

```python
h=2                                             #Tamaño del paso
a=0                                             #Limite inferior de Nuestro intervalo
b=50                                            #Limite superior de nuestro intervalo
m=72.4                                          #Masa del paracaidista
g=9.81                                          #Aceleracion gravedad de la tierra
c=13.2                                          #Coeficiente de arrastre.
N=(b-a)/h                                       #Calculamos el numero de iteraciones a partir del tamano de paso
i=1                                             #Nuestro contador de iteraciones
v=0                                             #Comenzando el movimiento desde 0    
t=a                                             #Establecemos el tiempo inicial como el limite inferior

print('{3}{0:^4}{3}{1:^8}{3}{2:^8}{3}'.format('i','t','T','|'))
while i<=N-1:
  plt.plot(t,v, '.r-')
  print('{3}{0:^4}{3}{1:>8.6f}{3}{2:>8.6f}{3}'.format(i,t,v,'|'))
  v=v+h*f(t,v)
  i=i+1
  t=a+i*h
  plt.xlabel('Tiempo (s)')
  plt.ylabel('Velocidad (m/s)')
  plt.ylim(0,100)
```
