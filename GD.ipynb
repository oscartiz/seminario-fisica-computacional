```python

def f1(t, y1, y2, y3, y4):

    return y2 

def f2(t, y1, y2, y3, y4):

    return 2*math.sin(y3)/(5+math.cos(y3))*y2*y4

def f3(t, y1, y2, y3, y4):

    return y4 

def f4(t, y1, y2, y3, y4):

    return -(5+math.cos(y3))*math.sin(y3)*y2**2 



#Iteramos RK4 numero de pasos
def rk4(t0, y1_0, y2_0, y3_0, y4_0, h,num_pasos):
  for i in range(num_pasos):
    t = t0
    y1 = y1_0
    y2 = y2_0
    y3 = y3_0
    y4 = y4_0
    k1_y1 = h * f1(t, y1, y2, y3, y4)
    k1_y2 = h * f2(t, y1, y2, y3, y4)
    k1_y3 = h * f3(t, y1, y2, y3, y4)
    k1_y4 = h * f4(t, y1, y2, y3, y4)

    k2_y1 = h * f1(t + h/2, y1 + k1_y1/2, y2 + k1_y2/2, y3 + k1_y3/2, y4 + k1_y4/2)
    k2_y2 = h * f2(t + h/2, y1 + k1_y1/2, y2 + k1_y2/2, y3 + k1_y3/2, y4 + k1_y4/2)
    k2_y3 = h * f3(t + h/2, y1 + k1_y1/2, y2 + k1_y2/2, y3 + k1_y3/2, y4 + k1_y4/2)
    k2_y4 = h * f4(t + h/2, y1 + k1_y1/2, y2 + k1_y2/2, y3 + k1_y3/2, y4 + k1_y4/2)

    k3_y1 = h * f1(t + h/2, y1 + k2_y1/2, y2 + k2_y2/2, y3 + k2_y3/2, y4 + k2_y4/2)
    k3_y2 = h * f2(t + h/2, y1 + k2_y1/2, y2 + k2_y2/2, y3 + k2_y3/2, y4 + k2_y4/2)
    k3_y3 = h * f3(t + h/2, y1 + k2_y1/2, y2 + k2_y2/2, y3 + k2_y3/2, y4 + k2_y4/2)
    k3_y4 = h * f4(t + h/2, y1 + k2_y1/2, y2 + k2_y2/2, y3 + k2_y3/2, y4 + k2_y4/2)

    k4_y1 = h * f1(t + h, y1 + k3_y1, y2 + k3_y2, y3 + k3_y3, y4 + k3_y4)
    k4_y2 = h * f2(t + h, y1 + k3_y1, y2 + k3_y2, y3 + k3_y3, y4 + k3_y4)
    k4_y3 = h * f3(t + h, y1 + k3_y1, y2 + k3_y2, y3 + k3_y3, y4 + k3_y4)
    k4_y4 = h * f4(t + h, y1 + k3_y1, y2 + k3_y2, y3 + k3_y3, y4 + k3_y4)

    y1 += (k1_y1 + 2*k2_y1 + 2*k3_y1 + k4_y1) / 6
    y2 += (k1_y2 + 2*k2_y2 + 2*k3_y2 + k4_y2) / 6
    y3 += (k1_y3 + 2*k2_y3 + 2*k3_y3 + k4_y3) / 6
    y4 += (k1_y4 + 2*k2_y4 + 2*k3_y4 + k4_y4) / 6

    t += h
    return y1, y2, y3, y4

#Interpolar vectores
def interpolar(vector1, vector2, t):
    interpolado = (1 - t) * vector1 + t * vector2
    return interpolado

def shooting_method(t0,velocidades1, velocidades2,h,y1,y3,num_pasos,num_interpol):
    alpha1=velocidades1[0]
    beta1=velocidades1[1]
    alpha2=velocidades2[0]
    beta2=velocidades2[1]
    y=rk4(t0, y1, alpha1, y3, beta1, h,num_pasos)
    vector1=np.array([y[1],y[3]])
    y=rk4(t0, y1, alpha2, y3, beta2, h,num_pasos)
    vector2=np.array([y[1],y[3]])
    vector3=interpolar(vector1, vector2, num_interpol)
    alpha3=vector3[0]
    beta3=vector3[1]
    return alpha3, beta3
```
